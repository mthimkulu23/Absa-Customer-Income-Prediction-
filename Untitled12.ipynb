```python
train['RECORD_DATE'] = pd.to_datetime(train['RECORD_DATE'])


train['year'] = train['RECORD_DATE'].dt.year
train['month'] = train['RECORD_DATE'].dt.month



def get_season(month):
    if month in [12, 1, 2]:
        return 'winter'
    elif month in [3, 4, 5]:
        return 'spring'
    elif month in [6, 7, 8]:
        return 'summer'
    else:
        return 'autumn'


train['season'] = train['month'].apply(get_season)

train
```

```python
# Stacking for Regression

from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingRegressor, RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
base_models = [
("lr", LinearRegression()),
("dt", DecisionTreeRegressor(random_state=42)),
("rf", RandomForestRegressor(n_estimators=100, random_state=42))
]
```

```python
meta_model = LinearRegression()
```

```python
stacking_model = StackingRegressor(
estimators=base_models,
final_estimator=meta_model,
cv=5
)

```

```python
stacking_model.fit(x_train, y_train)
```

```python
prediction_result = stacking_model.predict(x_test)
prediction_result
```

```python
accuracy = r2_score(y_test, prediction_result)
accuracy
```
